## 六、函数求导

### Sigmoid函数
\begin{equation*}
Sigmoid = 1/(1+\mathrm{e}^{-x}) 
\end{equation*}

\begin{equation*}
\frac{\partial (1/(1+\mathrm{e}^{-x}))}{\partial x} = 1/(1+\mathrm{e}^{-x}) \times (1- 1/(1+\mathrm{e}^{-x}))
\end{equation*}

\begin{equation*}
\frac{\partial Sigmoid}{\partial x} = Sigmoid \times (1- Sigmoid)
\end{equation*}


### Relu函数
\begin{equation*}
relu = max(0, x) 
\end{equation*}

\begin{equation*}
\ if x >0 , \frac{\partial (relu)}{\partial x} = 1
\end{equation*}

\begin{equation*}
\ Otherwise , \frac{\partial (relu)}{\partial x} = 0
\end{equation*}


### Softmax函数
\begin{equation*}
Softmax = \mathrm{e}^{x_{a}}/(\sum_{a=1}^{n}\mathrm{e}^{x_{a}}) = \mathrm{e}^{x_{1}}/(\mathrm{e}^{x_{1}} + \mathrm{e}^{x_{2}} + \mathrm{e}^{x_{3}}) 
\end{equation*}

\begin{equation*}
\frac{\partial (Softmax)}{\partial x_{1}} = (\mathrm{e}^{x_{1}} \times (\mathrm{e}^{x_{2}}+\mathrm{e}^{x_{3}}))/ (\mathrm{e}^{x_{1}}+\mathrm{e}^{x_{2}}+\mathrm{e}^{x_{3}})^2
\end{equation*}

- 通过数据对比，可以发现，层数越靠前，权值更新的幅度越小，这也就是神经网络在反向传播中梯度值逐渐减小的现象，也叫“梯度弥散”。

- 在早期，神经网络训练全部都是用sigmoid作为激励函数。当神经网络层数较多时，梯度值经过多层的传递后会呈现指数级急剧减小，在传递到前面几层时就会变得非常小。这种情况下，根据训练数据的反馈来更新神经网络的参数将会非常缓慢，基本起不到训练的作用。

- 直到ReLU的出现，才比较完美地解决了梯度弥散的问题。采用ReLU的原因：
    - Sigmoid在饱和区基本是一条水平线，导数约为0，由此得到的梯度也会非常小；而ReLU在大于0的时候，导数固定为1，这时梯度基本不会有什么损失，当然，在小于0的时候，导数也为0
（2）Sigmoid的导数涉及到指数和除法，计算量相对较大；而ReLU的导数计算则非常简单
（3）ReLU的激励结果在x<0的时候等于0，在前向计算的过程中，相当于Dropout，缓解了过拟合的问题。神经网络中解决过拟合问题常用的方法就是随机丢弃一些结点的值（Dropout），增加训练样本的多样性。
（4）ReLU函数非常类似于人脑的阈值响应机制。信号在超过某个阈值时，神经元才会进入兴奋和激活的状态，平时则处于抑制状态。

## 十一、偏差系数的更新

- 以其中一个偏差系数$b_{l1}$为例
$$\frac{\partial E_{total}}{\partial b_{l1}} = \frac{\partial E_{1}}{\partial O_{out1}} * \frac{\partial O
_{out1}}{\partial O_{in1}} * \frac{\partial O_{in1}}{\partial b_{l1}} = \frac{\partial E_{1}}{\partial O_{out1}} * \frac{\partial O_{out1}}{\partial O_{in1}} * 1$$

- 在形式上和更新矩阵系数类似，可以利用更新矩阵系数中间过程得到的数据

## 附：常见激励函数
    - sigmoid:$$sigmoid(z) = \frac{1}{1+exp(-z)}$$
    - hyperbolic tangent(双曲正切):$$tanh(z) = \frac{e^z -e^{-z}}{e^z + e^{-z}}$$
    - tanh(z)可以看成是sigmoid(z)的一个扩展：sigmoid(z)的输出范围是[0,1]，tanh(z)的输出范围是[-1, 1]
    - softplus:
    $$ softplus(z) = \log (1 + e^x)$$
    - ReLU:
    $$ relu(z) = max(0, x)$$

```python
import matplotlib.pyplot as plt
import numpy as np

xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)

plt.plot(xx, 1.0/(1+np.exp(-xx)), 'r-', label="sigmoid")
plt.plot(xx, (np.exp(xx) - np.exp(-xx))/(np.exp(xx) + np.exp(-xx)), 'c-', label="tanh")
plt.plot(xx, np.where(xx > 0, xx, 0), 'g-', label="ReLU")
plt.plot(xx, np.log(1+np.exp(xx)), 'm-', label="softplus")

plt.ylim(-2,2)
plt.legend(loc="lower right")
plt.xlabel(r"$z$")
plt.ylabel("$f(z)$")
plt.title('four kinds of activation function')
plt.show()
```
